```python
df1 = df.copy()

# strip + lowercase for safety
df1['Treatment Type'] = df['Treatment Type'].str.strip().str.lower()
df1['Early Diagnosis'] = df['Early Diagnosis'].str.strip().str.upper()
df1['Tobacco Use'] = df['Tobacco Use'].str.strip().str.upper()

# map to canonical labels
treat_map = {
    'chemo': 'Chemo',
    'no treatment': 'No Treatment',
    'surgery': 'Surgery',
    'radiation': 'Radiation',
    'targeted therapy': 'Targeted Therapy'
}
df1['treatment'] = df['Treatment Type'].map(treat_map)

df['early_dx'] = df['Early Diagnosis'].map({'Y':'Yes','N':'No'})
df['tobacco']  = df['Tobacco Use'].map({'Y':'Yes','N':'No'})

# stage as int, keep 0-4
df['stage'] = pd.to_numeric(df['Cancer Stage'], errors='coerce').astype('Int64')

# sanity checks
assert df['treatment'].isna().sum() == 0
assert df['early_dx'].isna().sum() == 0
assert df['tobacco'].isna().sum() == 0
```
